Aula 10 - Match Case

```python
# Peça ao usuário para digitar três lados de um triângulo e use match case para classificar o triângulo como:
# • Equilátero (todos os lados iguais)
# • Isósceles (dois lados iguais)
# • Escaleno (todos os lados diferentes)

lado1 = float(input("Digite o primeiro lado do triângulo: "))
lado2 = float(input("Digite o segundo lado do triângulo: "))
lado3 = float(input("Digite o terceiro lado do triângulo: "))

match (lado1 == lado2, lado1 == lado3 , lado2 == lado3):
  case (True, True, True):
    print("O Triângulo é Equilatero")
  case (True, False, False) | (False, True, False) | (False, False, True):
    print("O Triângulo é Isóceles")
  case _:
    print("O Triângulo é Escaleno")

```

```python
# Calculadora
num1 = int(input("Digite um número: "))
num2 = int(input("Digite outro número: "))
operacao = input("Digite o símbolo da operacão desejada: (+)Adicão, (-)Subtração, (*)Multiplicação ou (/)Divisão ").strip()

match (operacao):
  case "+":
    print(f"O resultado é {num1 + num2}")
  case "-":
    print(f"O resultado é {num1 - num2}")
  case "*":
    print(f"O resultado é {num1 * num2}")
  case "/":
    print(f"O resultado é {num1 / num2}")
  case _:
    print("Operador inválido.")
```

```python
#7. Ler duas notas de um aluno, efetuar a média aritmética e, se a média for maior ou igual a 7, informar que o aluno foi aprovado; se a média for maior ou igual a
# 5 mas menor do que 7, informar que o aluno está de exame; se a média for menor do que 5 informar que o aluno foi reprovado.
nota1 = float(input("Digite a primeira nota do aluno: "))
nota2 = float(input("Digite a segunda nota do aluno: "))

media = (nota1 + nota2) / 2
if media >= 7:
  print("O aluno foi aprovado.")
elif media >= 5 and media < 7:
  print("O aluno está de exame.")
else:
  print("O aluno foi reprovado.")
```

```python
# 8. Desenvolva um programa que recebe do usuário o placar de um jogo de futebol (os gols de cada time) e informe se o resultado foi um empate, se a vitória
# foi do primeiro time ou do segundo time.
golsTime1 = int(input("Digite a quantidade de gols do do primeiro time: "))
golsTime2 = int(input("Digite a quantidade de gols do do segundo time: "))

match (golsTime1 == golsTime2, golsTime1 > golsTime2, golsTime2 > golsTime1):
  case (True, False, False):
    print("Empate.")
  case (False, True, False):
    print("Vitória do primeiro time")
  case (False, False, True):
    print("Vitória do segundo time")
```

```python
#9. Faça um Programa que pergunte em que turno você estuda. Peça para digitar M matutino, V Vespertino ou N Noturno. Imprima a mensagem "Bom Dia!", "Boa
# Tarde!" ou "Boa Noite!" ou "Valor Inválido!", conforme o caso.
turno = input("Em qual turno você estuda? (M para matutino, V para Vespertino ou N para Noturno) ").upper().strip()

match turno:
  case "M":
    print("Bom Dia!")
  case "V":
    print("Boa Tarde!")
  case "N":
    print("Boa Noite!")
  case _:
    print("Valor Inválido!")
```

```python
# 11. Uma empresa resolveu dar um aumento de salário aos seus colaboradores e lhe contrataram para desenvolver o programa que calculará os reajustes. Faça
# um programa que recebe o salário de um colaborador e o reajuste segundo o seguinte critério, baseado no salário atual:
# salários até R$ 280,00 (incluindo): aumento de 20%
# salários entre R$ 280,00 e R$ 700,00: aumento de 15%
# salários entre R$ 700,00 e R$ 1500,00: aumento de 10%
# salários de R$ 1500,00 em diante: aumento de 5%

# Após o aumento ser realizado, informe na tela:
# o salário antes do reajuste;
# o percentual de aumento aplicado;
# o valor do aumento;
# o novo salário, após o aumento.
salario = float(input("Digite seu salário atual: "))

if salario <= 280:
  percentual = "20%"
  valorAumento = salario * 0.2
  reajuste = salario + valorAumento
elif salario >= 280 and salario < 700:
  percentual = "15%"
  valorAumento = salario * 0.15
  reajuste = salario + valorAumento
elif salario >= 700 and salario < 1500:
  percentual = "10%"
  valorAumento = salario * 0.1
  reajuste = salario + valorAumento
elif salario >= 1500:
  percentual = "5%"
  valorAumento = salario * 0.05
  reajuste = salario + valorAumento


print(f"Salário antes do reajuste: R$:{salario:.2f}")
print(f"Percentual do aumento aplicado: {percentual}")
print(f"Valor do aumento: R$:{valorAumento:.2f}")
print(f"Novo salário, após aumento: R$:{reajuste:.2f}")
```

```python
# 13. Faça um programa que pergunte ao usuário se ele quer passar uma temperatura de Fahrenheit para Celsius ou de Celsius para Fahrenheit, e que, a
# partir da resposta do usuário, faça a devida conversão.
conversao = input("Digite qual medida de temperatura voce quer converter ('F' de Fahrenheit para Celsius e 'C' de Celsius para Fahrenheit).").upper().strip()

if conversao == "F":
  fahrenheit = float(input("Digite a temperatura (em Fahrenheit): "))
  celsius = (fahrenheit - 32) / 1.8
  print(f"{fahrenheit} F°, em Celsius: {celsius} C°")
elif conversao == "C":
  celsius = float(input("Digite a temperatura (em Celsius): "))
  fahrenheit = (celsius * 1.8) + 32
  print(f"{celsius} C°, em Fahrenheit: {fahrenheit} F°")
else:
  print("Valor inválido.")
```
