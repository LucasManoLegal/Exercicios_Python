```python
#2. Crie um programa que leia a idade de uma pessoa e imprima:
# • "Criança" se a idade for menor que 12;
# • "Adolescente" se estiver entre 12 e 17;
# • "Adulto" se estiver entre 18 e 64;
# • "Idoso" se for 65 ou mais.

idade = int(input("Digite sua idade: "))
if idade < 12:
  print("Criança")
elif idade >= 12 and idade <= 17:
  print("Adolescente")
elif idade >=18 and idade <=64:
  print("Adulto")
elif idade >= 65:
  print("Idoso")
```

```python
#3. Desenvolva um programa que calcule o IMC (Índice de Massa Corporal) de uma pessoa e classifique-a como:
# • "Abaixo do peso" se IMC < 18.5;
# • "Peso normal" se 18.5 ≤ IMC < 25;
# • "Sobrepeso" se 25 ≤ IMC < 30;
# • "Obesidade" se IMC ≥ 30.
# • Receba o peso (kg) e a altura (m) do usuário.

peso = float(input("Digite seu peso: "))
altura = float(input("Digite sua altura: "))
imc = peso / (altura ** 2)
if imc < 18.5:
  print("Abaixo do peso")
elif imc >= 18.5 and imc < 25:
  print("Peso normal")
elif imc >= 25 and imc < 30:
  print("Sobrepeso")
elif imc >= 30:
  print("Obesidade")
```

```python
#4. Crie um programa que calcule o bônus de um funcionário com base
# no tempo de serviço:
# • Sem bônus para menos de 1 ano;
# • Bônus de 5% para 1 a 3 anos;
# • Bônus de 10% para 3 a 5 anos;
# • Bônus de 15% para mais de 5 anos.
# • Solicite o salário e o tempo de serviço, e exiba o salário final com
# bônus (se aplicável).

salario = float(input("Digite seu salário: "))
tempoServico = int(input("Digite seu tempo de serviço (em anos): "))
if tempoServico < 1:
  print("Sem Bônus.")
elif tempoServico >= 1 and tempoServico < 3:
  salario += salario * 0.05
  print(f"Bônus de 5% concedido. valor final {salario:.2f}")
elif tempoServico >= 3 and tempoServico <= 5:
  salario += salario * 0.1
  print(f"Bônus de 10% concedido. valor final {salario:.2f}")
elif tempoServico < 5:
  salario += salario * 0.15
  print(f"Bônus de 15% concedido. valor final {salario:.2f}")
```

```python
# 5. Escreva um programa que leia a velocidade de um veículo e a
# classifique em:
# • "Baixa velocidade" para velocidades até 40 km/h;
# • "Velocidade moderada" para velocidades entre 41 e 80 km/h;
# • "Velocidade alta" para velocidades entre 81 e 120 km/h;
# • "Velocidade muito alta" para velocidades acima de 120 km/h.

velocidade = float(input("Digite a velocidade do veículo em km/h: "))
if velocidade <= 40:
    print("Baixa velocidade")
elif 41 <= velocidade <= 80:
    print("Velocidade moderada")
elif 81 <= velocidade <= 120:
    print("Velocidade alta")
else:
    print("Velocidade muito alta")
```
