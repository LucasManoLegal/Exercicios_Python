```python
# Imprimir Olá, Mundo! 5 vezes
# Escreva um programa que utilize um laço for para imprimir a mensagem
# Olá, Mundo! cinco vezes na tela.

for i in range(5):
  print("Olá, Mundo!")
```

```python
# Contar de 1 a 50
# Utilize um laço for para exibir todos os números de 1 a 50.

for i in range(1, 51):
  print(i)
```

```python
# Exibir números ímpares de 1 a 20
# Escreva um programa que exiba apenas os números ímpares de 1 a 20.

for i in range(1, 21, 2):
  print(i)
```

```python
# Multiplicar cada número de uma lista por 2
# Dada uma lista [1, 2, 3, 4, 5], utilize um laço for para imprimir cada número
# multiplicado por 2.

lista = [1, 2, 3, 4, 5]

for i in lista:
  print(i*2)
```

```python
# Imprimir os caracteres de uma palavra separadamente
# Peça ao usuário para digitar uma palavra e, utilizando um for, exiba cada
# caractere separadamente.

palavra = input("Digite uma palavra: ")
for i in palavra:
  print(i)
```

```python
# Contagem de 10 até 1
# Escreva um programa que exiba uma contagem regressiva de 10 até 1.

for i in range(10, 0, -1):
  print(i)
```

```python
# Somar os números de 1 a 10
# Utilize um laço for para calcular e exibir a soma dos números de 1 a 10.
soma = 0
for i in range(10):
  soma += i + 1
print(soma)
```

```python
# Pedir 5 números ao usuário e exibi-los
# Peça para o usuário digitar 5 números e, em seguida, exiba cada um
# deles.

num = ""
for i in range(5):
  num = input("Digite um número: ")
  print(num)
```

```python
# Exibir a tabuada do 7
# Use um laço for para exibir a tabuada do 7 (de 1 a 10).

print("Tabuada do 7")
for i in range(1, 11):
  print(f"7 x {i} = {7*i}")
```

Estruturas dentro de estruturas

```python
# for dentro de for
print("Tabuada completa")
for i in range(1, 11):
  for j in range(1, 11):
    print(f"{i} x {j} = {i*j}")
```

```python
# vogais e consoantes
palavra = input("Digite uma palavra: ").lower()
vogais = "aeiou"
qtdVogais = 0
qtdConsoantes = 0
qtdLetras = len(palavra)
for i in palavra:
  if i in vogais:
    qtdVogais += 1
  else:
    qtdConsoantes += 1

print(f"A palavra {palavra} tem {qtdLetras} letras, sendo {qtdConsoantes} consoantes e {qtdVogais} vogais.")
```
