```python
#1. Verificação de números diferentes
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))
if num1 != num2:
  print("Os números são diferentes")
else:
  print("Os numeros são iguais")
```

```python
#2. Verificação de maioridade
idade = int(input("Digite a sua idade: "))
if idade >= 18:
  print("Você é maior de idade")
else:
  print("Você é menor de idade")
```

```python
#3. Autorização de entrada em um evento
senhaConvite = input("Digite a senha do convite: ")
if senhaConvite == "1234":
  print("Entrada permitida")
else:
  print("Entrada negada")
```

```python
#4. Verificação de nome em lista
nome = input("Digite seu nome: ").upper().strip()
if nome == "CARLOS":
  print("Você está na lista")
else:
  print("Nome não encontrado")
```

```python
#5. Cálculo de gorjeta
valorConta = float(input("Digite o valor da conta: R$:"))
if valorConta > 100.00:
  valorConta += 0.1 * valorConta
  print(f"Total a pagar: R$:{valorConta:.2f}")
else:
  valorConta += 0.05 * valorConta
  print(f"Total a pagar: R$:{valorConta:.2f}")
```

```python
#6 Acesso ao Wi-Fi
senha = input("Digite a senha do wi-fi: ")
if senha == "senha123":
  print("Conectado")
else:
  print("Senha incorreta")
```

```python
#7. Verificação de turno
turno = input("Digite no espaço a seguir conforme a instrução: (M para manhã, qualquer tecla para tarde):").upper().strip()
if turno == "M":
  print("Bom dia!")
else:
  print("Boa tarde!")
```

```python
#8. Verificação de múltiplo de 5
num = int(input("Digite um número: "))
if num % 5 == 0:
  print("Multiplo de 5")
else:
  print("Não é múltiplo de 5")
```

```python
#9. Desconto em compras por valor mínimo
valorCompra = float(input("Digite o valor da compra: R$:"))
if valorCompra > 150.00:
  valorCompra -= 20
  print(f"Valor final: R$:{valorCompra}")
else:
  print(f"Valor final: R$:{valorCompra}")
```
